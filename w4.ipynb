```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import scipy.io as sio
import scipy.optimize as opt
import matplotlib
from sklearn.metrics import classification_report   #这个包是评价报告

'''1。载入mat文件，读取数据'''
##############################################################################
########---------------------打开mat 文件--------------------------########
##############################################################################
def load_data(path, transpose=True):
    '''
    这是python读取mat文件的方法：
    data = sio.loadmat('/Users/jiayun/PycharmProjects/untitled/'
                       'untitled/untitled/rbot Project/ my code/machine learning/w'
                       'eek4 neural network/ex3data1.mat')
    p ='/Users/jiayun/PycharmProjects/untitled/'
                       'untitled/untitled/rbot Project/ my code/machine learning/w'
                       'eek4 neural network/ex3data1.mat'
    '''
    data = sio.loadmat(path)
    print(sio.whosmat(path))  # 这个命令可以给出mat文件的变量信息
    y = data.get('y')
    y = y.reshape(y.shape[0])

    x = data.get('X')
    if transpose:
        x = np.array([im.reshape((20,20)).T for im in x])

        x = np.array([im.reshape(400) for im in x])

    return x, y



x, y = load_data('ex3data1.mat')
print(x.shape)
print(y.shape)

# plot an image
def plot_an_image(image):
     """
     image : (400,)
     """
     fig,ax = plt.subplots(figsize=(1,1))
     ax.matshow(image.reshape((20,20)),cmap =matplotlib.cm.binary)
     plt.xticks(np.array([]))
     plt.yticks(np.array([]))

# 随机选一个。画出来看看
pick_one = np.random.randint(0,5000)      # 在（0，5000）中随机选一个
plot_an_image(x[pick_one,:])
plt.show()
print('this picture should belong to class {}'.format(y[pick_one]))


##############################################################################
########---------------------打开mat 文件 end--------------------------########
##############################################################################

##############################################################################
########---------------------打开mat 文件--------------------------########
##############################################################################
```
